```python
# Creating a Class to get student info and to calculate their avg_marks
class Student: 
    def __init__(self, name, marks):  # Taking info of student in parameters
        self.name = name              # object name stored 
        self.marks = marks            # object marks stored

    def avg_marks(self):              # Created a function to calculate average marks 
        sum = 0                       # to store objects marks
        count = 0                     # to calculate how many subjects marks, object are initiating 
        for val in self.marks:        # to get the every value of each subject marks
            sum += val                # adding each subject val and sum all the val
            count += 1                # total subject count
        print(f"Hi {self.name}, Your Average Marks is {round((sum/count), 2)}")    # sum val/the number of subject and rounding decimal val of avg marks into 2 decimal point

# Making Object 
s1 = Student("Sanu", [90, 90, 80])
s2 = Student("Mussu", [94, 79, 69])
s3 = Student("Shubham", [88, 70, 95])

# Calling Avg_marks Function/Method 
s1.avg_marks()
s2.avg_marks()
s3.avg_marks()
```

```python
# Using staticmethod and inheritance
class Car:                # Parent Class
    def __init__(self, type):
        self.type = type
        
    @staticmethod         # Staticmethod: works independently, doesn't use class or object data
    def start():
        print('Car Started!')

    @staticmethod         # Staticmethod: works independently, doesn't use class or object data
    def stop():
        print('Car Stoped!')

class ToyotaCar(Car):     # Used Inheritance concept to inherit the features of parent class (Car)
    def __init__(self, name, type):
        self.name = name
        super().__init__(type)
        super().start()


# Create an object
car1 = ToyotaCar('Fortuner', 'Diesel')
print(f'{car1.name}\n{car1.type}')


# Output:
# __________
# Car Started!    # got this output because when we create an object, '__init__' runs automatically with all defined func. (super.start()) 
# Fortuner
# Diesel
```

```python
# @property method : 
# -----------------
# What is a property method?
# @property is a Python decorator that lets you use a method like an attribute.
# It’s mainly used to control access to class data (variables).

# Why we use it:
# Encapsulation → hide and protect internal data.
# Validation → check or restrict values before setting.
# Cleaner syntax → access methods like simple attributes (no () needed).
# Read-only / computed attributes → let user read but not modify.

class Student: 
    def __init__(self, phy, chem, math):
        self.phy = phy
        self.chem = chem
        self.math = math

    @property
    def percentage(self): 
        print((self.phy + self.chem + self.math) / 3, "%") 

# Creating Object
stud1 = Student(98, 97, 99)
stud1.percentage    # Using @property doesn't required Parathises to call the func

# Updating marks
stud1.phy = 86     
stud1.percentage
```

```python
# Polymorphism: Dunder Function 

# Complex Number Function
class Complex: 
    def __init__(self, real, img, from_add=False):
        self.real = real
        self.img = img
        self.from_add = from_add   # flag to identify if created by add()

    def showNumber(self):
        if self.from_add:   # condition check
            print(f'{self.real}i + {self.img}j')
        else:
            pass  # do nothing if not created by add()

    def add(self, num2): 
        newReal = self.real + num2.real
        newImg = self.img + num2.img
        # while returning new object, mark it as "created by add()"
        return Complex(newReal, newImg, from_add=True)   # returning the operation value (addition val) to Class 'Complex' parameters 'real, img'


# Creating Objects 
num1 = Complex(1, 3)
num1.showNumber()  # nothing prints (not from add)

num2 = Complex(4, 6)
num2.showNumber()  # nothing prints (not from add)

num3 = num1.add(num2)
num3.showNumber()  # prints only this one
```
