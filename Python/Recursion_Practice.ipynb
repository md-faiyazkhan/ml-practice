### Recursion is a process in which a function calls itself until a base condition is met.

```python
# Factorial Number
def factorial(n): 
    if(n==1 or n==0): 
        return 1
    return n * factorial(n-1)
    
n = int(input("Enter a number: "))   
print(f"The factorial of this number is: {factorial(n)}")
```
