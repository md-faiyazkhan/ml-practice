```python
blogHeading = 'python is perfect choise for Ai.'
print(blogHeading.title())  # to capitalize first later of each word from a sentance
print(blogHeading.upper())  # to capitalize all later
print(blogHeading.lower())  # to small laters for all laters
print(blogHeading.capitalize())  # capitalize only first later from a sentance

# If by mistake you right string like this in your variable
blog = 'pytHOn is a Slow langUage'
print(blog.capitalize())  # it will fix your string data. First later will be capital and rest later will become small laters
```

```python
name = '!!!!faiyaz!!!!!!!'
print(name.strip('!'))  # remove mention word (!) from string
print(name.lstrip('!'))  # lstrip = left strip
print(name.rstrip('!'))  # rstrip = right strip
```

```python
name = '!!!!faiyaz!!!!!!!'
name1 = '!!!!faiyaz!!!!!!!faiyaz'
print(name.replace('faiyaz', 'sanu'))  # replace all occurance word with the second mention word
print(name1.replace('faiyaz', 'sanu'))  # replace all occurance word with the second mention word
```

```python
text = "Python ek powerful language hai"
words = text.split()
print(words) 
# Output: ['Python', 'ek', 'powerful', 'language', 'hai']
```

```python
data = "Apple,Banana,Mango,Orange"
fruits = data.split(",")
print(fruits)
# Output: ['Apple', 'Banana', 'Mango', 'Orange']
```

```python
text = "one:two:three:four"
# Sirf 2 baar split karega
result = text.split(":", 2)
print(result)
# Output: ['one', 'two', 'three:four']
```
