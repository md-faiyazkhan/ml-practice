```python
# Take an input from user and find the length of words of given user inpput
def count_words():
    sentence = input("Enter a sentence : ")
    return len(sentence)

# Function Call 
count_words()
```

```python
# Take an input from user and find the length of words of given user inpput
def counting(sentence):   # sentence is an parameter which need some arguments. Parameter=Variable, Arguments=Value
    sentence = input("Enter a Sentence: ")
    count = 0
    for char in sentence:
        if char == ' ':
            count += 1
    return count     # 'return' is returning the operation calculation value to its Parameter='Sentence'

# Function Call 
print("No. of words:", counting(sentence))    # 'counting' is calling the function and 'sentence' is passing the functions value
```

```python
def goodDay(name, ending="Thank You"):
    print(f"Good Day, {name}")   # name is an Parameter which need some Arguments. Parameter=Variable, Arguments=Value
    print(ending)
    
goodDay("Sanu")  # Passing Argument while calling Function
goodDay("Sanu", "Thanks")   # "Thanks" will be the Argument of 'ending' parameter. 'Thank You' Argument will work as Default, when i don't provide Argument manully.
```

```python
# Q. Create a Multiplication table
def multiplication(): 
    n = int(input("Please Enter a Number: "))
    for i in range(1, 11): 
        print(f"{n} X {i} = {n*i}")

multiplication()
```
