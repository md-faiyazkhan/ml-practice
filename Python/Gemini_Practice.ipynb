```python
# Problem scenario: Hamare paas multiple cities me branches hai electronics ki or hame ye check karna hai apne data se ki kis kis cities me currently laptop available hai. 
# Dictionary 
# 1. Aapka Original Data (Slow Search)
inventory = {
    "Delhi": ["Mobile", "Laptop", "Tablet"],
    "Mumbai": ["Mobile", "Charger", "Laptop"],
    "Bangalore": ["Tablet", "Camera", "Charger"]
}

# 2. Step 1: Data ko "Ulatna" (Reverse Mapping)
# Hum ek baar loop chalakar fast_lookup dictionary banayenge
fast_lookup = {}

for city, items in inventory.items():
    for item in items:
        if item not in fast_lookup:
            fast_lookup[item] = [city] # Naya item mila toh city ki list shuru karo
        else:
            fast_lookup[item].append(city) # Pehle se hai toh sirf city jodo

# 3. Step 2: Ab bina loop ke "Direct Search" (The Magic Line)
# Ab aap 50,000 cities bhi dalenge, ye line instantly result degi
print(fast_lookup["Laptop"]) 
# Output: ['Delhi', 'Mumbai']
```
