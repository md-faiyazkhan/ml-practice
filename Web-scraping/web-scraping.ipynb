<div> 
    <h1 style="color: #FFD700;">  Creating DataFrame for all the Pages  </h1>
    <p style = "margin-top: -15px; margin-left: 5px; opacity: 0.8;"> Web Scraped Data Output </p> 
</div>

```python
# 1. Create an empty list at the very outside (to add dataframes)
all_pages_data = []

for j in range(1, 1001): 

    url = f'https://www.ambitionbox.com/list-of-companies?page={j}'

    # Define headers properly
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    
    # Send request
    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    company = soup.find_all('div', class_='companyCardWrapper')

    name, rating, reviews, ctype, hq = [], [], [], [], []
    
    
    for i in company:
        try:
            name.append(i.find('h2').text.strip())
        except: name.append(np.nan)

        try:
            # FIX 3: .split() + .join() logic for cleaner output
            r = i.find('div', attrs={"style": "height:auto;padding-bottom:1px;"}).text
            rating.append(" ".join(r.split()))
        except: rating.append(np.nan) # Spelling fix: appned -> append

        try:
            rev = i.find('span', class_='companyCardWrapper__companyRatingCount').text
            reviews.append(" ".join(rev.split()))
        except: reviews.append(np.nan)

        try:
            interlinking = i.find('span', class_='companyCardWrapper__interLinking').text.split('|')
            ctype.append(interlinking[0].strip())
            hq.append(interlinking[1].strip() if len(interlinking) > 1 else np.nan)
        except:
            ctype.append(np.nan)
            hq.append(np.nan)
    
    
    
    # Current page data dictionary
    d = {'name': name, 'rating': rating, 'reviews': reviews, 'company_type': ctype, 'Head_Quarters': hq}
    
    # Create dataframes inside the loop and add them to the list
    df = pd.DataFrame(d)
    all_pages_data.append(df)

# Add them all together after the loop ends (Anti-AttributeError)
final = pd.concat(all_pages_data, ignore_index=True)
```
