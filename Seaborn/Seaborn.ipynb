```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
penguins = sns.load_dataset('penguins')
```

```python
# Show first 5 rows
penguins.head()
```

```python
# Create scatter plot
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species', style='sex')

# Add title and labels
plt.title('Penguin Bill Length vs Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')

# Show the plot
plt.show()
```

```python
# Bar plot of average flipper length per species
plt.figure(figsize=(8,5))

sns.barplot(data=penguins, x='species', y='flipper_length_mm', ci=None, palette='pastel')

# Add title and labels
plt.title('Average Flipper Length per Penguin Species')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')

# Show the plot
plt.show()
```

```python
# Scatter plot with regression line: body mass vs flipper length
plt.figure(figsize=(8,5))

sns.regplot(data=penguins, x='flipper_length_mm', y='body_mass_g', scatter_kws={'s':50}, line_kws={'color':'red'})

# Add title and labels
plt.title('Body Mass vs Flipper Length with Regression Line')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')

# Show the plot
plt.show()
```

```python
# Box plot of body mass per species
plt.figure(figsize=(8,5))

sns.boxplot(data=penguins, x='species', y='body_mass_g', palette='Set2')

# Add title and labels
plt.title('Body Mass Distribution per Penguin Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')

# Show the plot
plt.show()
```

```python
# Violin plot of body mass by species and island
plt.figure(figsize=(10,6))

sns.violinplot(data=penguins, x='species', y='body_mass_g', hue='island', split=True, palette='muted')

# Add title and labels
plt.title('Penguin Body Mass Distribution by Species and Island')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')

# Show the plot
plt.show()
```

```python
# Correlation heatmap of numerical features in penguins dataset
plt.figure(figsize=(8,6))

# Drop non-numeric columns before correlation
numeric_data = penguins.drop(columns=['species','island','sex'])

sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")

# Add title
plt.title('Correlation Heatmap of Penguin Features')

# Show the plot
plt.show()
```

```python
# Pairplot of penguins dataset
sns.pairplot(data=penguins, hue='species', diag_kind='kde', palette='Set1')

# Show the plot
plt.show()
```

```python
# Countplot of penguin species per island
plt.figure(figsize=(8,5))

sns.countplot(data=penguins, x='island', hue='species', palette='Set3')

# Add title and labels
plt.title('Count of Penguin Species per Island')
plt.xlabel('Island')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
# Sex-wise body mass distribution using boxenplot
plt.figure(figsize=(8,5))

sns.boxenplot(data=penguins, x='sex', y='body_mass_g', hue='species', palette='pastel')

# Add title and labels
plt.title('Sex-wise Body Mass Distribution by Species')
plt.xlabel('Sex')
plt.ylabel('Body Mass (g)')

# Show the plot
plt.show()
```

```python
# Swarmplot of body mass by species
plt.figure(figsize=(8,5))

sns.swarmplot(data=penguins, x='species', y='body_mass_g', hue='sex', palette='Set2')

# Add title and labels
plt.title('Penguin Body Mass by Species (Swarmplot)')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')

# Show the plot
plt.show()
```

```python
# FacetGrid: Flipper length distribution per species
g = sns.FacetGrid(penguins, col="species", height=4, aspect=1.2)
g.map(sns.histplot, "flipper_length_mm", bins=20, kde=True, color="skyblue")

# Add titles
g.set_axis_labels("Flipper Length (mm)", "Count")
g.set_titles("Species: {col_name}")

plt.show()
```

```python
# Jointplot of bill length vs bill depth with species hue
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="scatter", palette="Dark2")

plt.show()
```

```python
# 2D KDE plot of bill length vs bill depth
plt.figure(figsize=(7,5))

sns.kdeplot(
    data=penguins, 
    x="bill_length_mm", 
    y="bill_depth_mm", 
    fill=True, 
    cmap="mako", 
    thresh=0.05, 
    levels=15
)

# Add title and labels
plt.title("2D KDE Density Plot: Bill Length vs Bill Depth")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")

plt.show()
```

```python
# Stripplot of flipper length by species
plt.figure(figsize=(8,5))

sns.stripplot(data=penguins, x="species", y="flipper_length_mm", hue="sex", jitter=True, dodge=True, palette="Set1")

# Add title and labels
plt.title("Stripplot: Flipper Length by Species and Sex")
plt.xlabel("Species")
plt.ylabel("Flipper Length (mm)")

plt.show()
```

```python
# Lineplot of average body mass vs flipper length
plt.figure(figsize=(8,5))

sns.lineplot(
    data=penguins, 
    x="flipper_length_mm", 
    y="body_mass_g", 
    hue="species", 
    estimator="mean", 
    ci=None, 
    marker="o"
)

# Add title and labels
plt.title("Lineplot: Avg Body Mass vs Flipper Length by Species")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")

plt.show()
```
