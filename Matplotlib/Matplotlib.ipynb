```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
from sklearn.datasets import load_wine
```

```python
# Load dataset
wine = load_wine()
```

```python
# Convert to DataFrame
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['target'] = wine.target

# First look at data
print(df.head())
```

```python
# 1. Basic Histogram of 'alcohol'
plt.figure(figsize=(6,4))
plt.hist(df['alcohol'], bins=20, color='skyblue', edgecolor='black')
plt.title("Histogram of Alcohol")
plt.xlabel("Alcohol")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Scatter plot (alcohol vs malic_acid)
plt.figure(figsize=(6,4))
plt.scatter(df['alcohol'], df['malic_acid'], c=df['target'], cmap='viridis', edgecolor='k')
plt.title("Alcohol vs Malic Acid")
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.colorbar(label="Wine Class")
plt.show()
```

```python
# Q: Scatter plot between Alcohol and Malic Acid

plt.figure(figsize=(6,4))
plt.scatter(df['alcohol'], df['malic_acid'], color="tomato", alpha=0.7, edgecolors="k")
plt.title("Alcohol vs Malic Acid")
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.show()
```

```python
# Q: Boxplot of Alcohol

plt.figure(figsize=(5,4))
plt.boxplot(df['alcohol'], patch_artist=True, boxprops=dict(facecolor="skyblue"))
plt.title("Boxplot of Alcohol")
plt.ylabel("Alcohol")
plt.show()
```

```python
# Q: Scatter plot of Alcohol vs Malic Acid

plt.figure(figsize=(6,5))
plt.scatter(df['alcohol'], df['malic_acid'], c='purple', alpha=0.6, edgecolors='k')
plt.title("Alcohol vs Malic Acid")
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.show()
```

```python
# Q: Plot boxplot of 'alcohol' feature from wine dataset to detect outliers.

plt.figure(figsize=(6,4))
plt.boxplot(df['alcohol'])
plt.title("Boxplot of Alcohol")
plt.ylabel("Alcohol")
plt.show()
```

```python
# Q: Plot scatter plot between 'alcohol' and 'malic_acid' to see relationship.

plt.figure(figsize=(6,4))
plt.scatter(df['alcohol'], df['malic_acid'], alpha=0.6)
plt.title("Scatter Plot: Alcohol vs Malic Acid")
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.show()
```

```python
# Q: Plot boxplot of 'alcohol' feature to check distribution and outliers.

plt.figure(figsize=(6,4))
plt.boxplot(df['alcohol'], vert=True, patch_artist=True)
plt.title("Boxplot of Alcohol")
plt.ylabel("Alcohol")
plt.show()
```

```python
# Q: Plot scatter plot between 'alcohol' and 'malic_acid' to observe correlation.

plt.figure(figsize=(6,4))
plt.scatter(df['alcohol'], df['malic_acid'], alpha=0.6, edgecolor='k')
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.title("Scatter Plot: Alcohol vs Malic Acid")
plt.show()
```

```python
# Q: Plot boxplot of 'alcohol' feature to visualize distribution and outliers.

plt.figure(figsize=(6,4))
plt.boxplot(df['alcohol'], vert=False)
plt.xlabel("Alcohol")
plt.title("Boxplot of Alcohol")
plt.show()
```

```python
# Q: Plot scatter plot between 'alcohol' and 'malic_acid' to see relationship.

plt.figure(figsize=(6,4))
plt.scatter(df['alcohol'], df['malic_acid'], alpha=0.7)
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.title("Scatter Plot of Alcohol vs Malic Acid")
plt.show()
```

```python
# Q: Plot scatter plot between 'alcohol' and 'malic_acid' to see relationship.

plt.figure(figsize=(6,4))
plt.scatter(df['alcohol'], df['malic_acid'], alpha=0.7)
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.title("Scatter Plot of Alcohol vs Malic Acid")
plt.show()
```

```python
# Q: Plot pairwise scatter plots of selected features from wine dataset.

pd.plotting.scatter_matrix(
    df[['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash']], 
    figsize=(10, 8), 
    diagonal='hist', 
    color='teal', 
    alpha=0.7
)
plt.suptitle("Pairwise Scatter Matrix of Selected Features", y=1.02)
plt.show()
```

```python
# Q: Plot a stacked bar chart showing mean values of selected features grouped by target.

grouped_means = df.groupby(wine.target)[['alcohol', 'malic_acid', 'ash']].mean()

grouped_means.plot(
    kind='bar', 
    stacked=True, 
    figsize=(8, 6), 
    colormap='viridis'
)

plt.title("Stacked Bar Chart of Mean Feature Values by Wine Class")
plt.xlabel("Wine Class")
plt.ylabel("Mean Value")
plt.legend(title="Features")
plt.show()
```

```python
# Q: Plot line chart of 'alcohol' values for first 50 wine samples.

plt.figure(figsize=(10,5))
plt.plot(df['alcohol'][:50], marker='o', linestyle='-', color='navy')
plt.title("Line Plot of Alcohol (First 50 Samples)")
plt.xlabel("Sample Index")
plt.ylabel("Alcohol")
plt.show()
```

```python
# Q: Plot histogram of 'malic_acid' feature to observe distribution.

plt.figure(figsize=(6,4))
plt.hist(df['malic_acid'], bins=20, color='orange', edgecolor='black', alpha=0.7)
plt.title("Histogram of Malic Acid")
plt.xlabel("Malic Acid")
plt.ylabel("Frequency")
plt.show()
```

```python
# Q: Plot bar chart of average alcohol content per wine class.

avg_alcohol = df.groupby(wine.target)['alcohol'].mean()

plt.figure(figsize=(6,4))
plt.bar(avg_alcohol.index, avg_alcohol.values, color='green', alpha=0.7)
plt.title("Average Alcohol Content per Wine Class")
plt.xlabel("Wine Class")
plt.ylabel("Average Alcohol")
plt.show()
```

```python
# Q: Plot density curve (KDE) of 'alcohol' feature.

plt.figure(figsize=(6,4))
df['alcohol'].plot(kind='kde', color='red', linewidth=2)
plt.title("Density Plot of Alcohol")
plt.xlabel("Alcohol")
plt.show()
```

```python
# Q: Plot stacked area chart of mean values of selected features by wine class.

mean_features = df.groupby(wine.target)[['alcohol', 'malic_acid', 'ash']].mean()

plt.figure(figsize=(8,6))
plt.stackplot(mean_features.index, 
              mean_features['alcohol'], 
              mean_features['malic_acid'], 
              mean_features['ash'], 
              labels=['Alcohol', 'Malic Acid', 'Ash'], 
              alpha=0.8)

plt.title("Stacked Area Chart of Mean Feature Values by Wine Class")
plt.xlabel("Wine Class")
plt.ylabel("Mean Value")
plt.legend(loc='upper left')
plt.show()
```

```python
# Q: Plot pie chart of wine class distribution.

class_counts = df.groupby(wine.target).size()

plt.figure(figsize=(6,6))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title("Wine Class Distribution")
plt.show()
```

```python
# Q: Plot horizontal bar chart of average malic acid per wine class.

avg_malic = df.groupby(wine.target)['malic_acid'].mean()

plt.figure(figsize=(6,4))
plt.barh(avg_malic.index, avg_malic.values, color='purple', alpha=0.7)
plt.title("Average Malic Acid per Wine Class")
plt.xlabel("Average Malic Acid")
plt.ylabel("Wine Class")
plt.show()
```

```python
# Q: Plot cumulative distribution of 'alcohol' feature.

plt.figure(figsize=(6,4))
plt.hist(df['alcohol'], bins=30, cumulative=True, density=True, color='skyblue', edgecolor='black')
plt.title("Cumulative Distribution of Alcohol")
plt.xlabel("Alcohol")
plt.ylabel("Cumulative Probability")
plt.show()
```

```python
# Q: Plot error bar chart of alcohol mean with standard deviation for each wine class.

import numpy as np

alcohol_mean = df.groupby(wine.target)['alcohol'].mean()
alcohol_std = df.groupby(wine.target)['alcohol'].std()

plt.figure(figsize=(6,4))
plt.errorbar(alcohol_mean.index, alcohol_mean.values, yerr=alcohol_std.values, fmt='o', capsize=5, color='darkblue')
plt.title("Alcohol Mean with Standard Deviation per Wine Class")
plt.xlabel("Wine Class")
plt.ylabel("Alcohol")
plt.show()
```

```python
# Q: Plot stacked histogram of 'alcohol' feature for each wine class.

plt.figure(figsize=(8,6))
plt.hist([df['alcohol'][wine.target==0],
          df['alcohol'][wine.target==1],
          df['alcohol'][wine.target==2]],
         bins=20,
         stacked=True,
         label=['Class 0','Class 1','Class 2'],
         alpha=0.7)

plt.title("Stacked Histogram of Alcohol by Wine Class")
plt.xlabel("Alcohol")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Q: Plot hexbin plot of 'alcohol' vs 'malic_acid'.

plt.figure(figsize=(7,5))
plt.hexbin(df['alcohol'], df['malic_acid'], gridsize=30, cmap='Blues')
plt.colorbar(label="Counts")
plt.title("Hexbin Plot: Alcohol vs Malic Acid")
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.show()
```

```python
# Q: Plot violin plot of 'alcohol' distribution per wine class.

import seaborn as sns

plt.figure(figsize=(7,5))
sns.violinplot(
    data=df.assign(target=wine.target),
    x="target",
    y="alcohol",
    hue="target",          # hue add kiya
    palette="Set2",
    legend=False
)
plt.title("Violin Plot of Alcohol by Wine Class")
plt.xlabel("Wine Class")
plt.ylabel("Alcohol")
plt.show()
```

```python
# Q: Plot stem plot of alcohol values (first 30 samples).

plt.figure(figsize=(8,4))
markerline, stemlines, baseline = plt.stem(df['alcohol'][:30], basefmt=" ")
plt.setp(markerline, color="red", marker="o")
plt.setp(stemlines, color="blue")
plt.title("Stem Plot of Alcohol (First 30 Samples)")
plt.xlabel("Sample Index")
plt.ylabel("Alcohol")
plt.show()
```
