```python
import numpy as np 
```

```python
# Creating an array 
arr1 = np.array([2, 6, 3, 9, 4, 2, 7, 4, 1, 9, 8])
print("Original arr1:", arr1)
```

```python
# arr1 ke same shape ka zeros array banana
zeros = np.zeros(arr1.shape)
print("Zeros array:", zeros)
```

```python
ones = np.ones(arr1.shape)
print("Ones array:", ones)
```

```python
full = np.full(arr1.shape, 12)
print("filled value:", full)
```

```python
full = np.full((4,), 12)
np.full(arr1.shape, 12)
print("filled value:", full) 
```
