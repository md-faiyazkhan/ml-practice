```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.sample(5)  # random 5 data
```

# 1. Categorical Data

# a. Countplot

```python
# using seaborn only
sns.countplot(x='Survived', data=df)
```

```python
# using pandas and seaborn 
df['Survived'].value_counts().plot(kind='bar')
```

# b. PieChart

```python
# using pandas and seaborn  
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
plt.show()
```

# 2. Numerical Data

# a. Histogram

```python
plt.hist(df['Age'])
```

# b. Distplot

```python
sns.distplot(df['Age'])  # older version will be removed from seaborn
```

```python
# cut=5 line ko floor tak le jayega, alpha=0.4 use karein purane look ke liye
sns.histplot(df['Age'], kde=True, kde_kws={'cut': 3}, edgecolor=None, alpha=0.4, stat="density")

plt.show()
```

# c. Boxplot

```python
sns.boxplot(x='Age', data=df)
```
