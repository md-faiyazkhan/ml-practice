```python
import numpy as np 
from scipy import stats 
```

```python
group_A = [85, 88, 90, 92, 87, 85, 89, 91, 86, 88] 
group_B = [82, 84, 80, 83, 81, 79, 78, 85, 84, 83]  
```

```python
t_stats , p_value = stats.ttest_ind(group_A, group_B , equal_var= False)    # ttest.ind = 2 tailed test. Mostly we preffer to use equal_varience=False
```

```python
t_stats
```

```python
p_value
```

```python
alpha = 0.05   # significance Value
```

```python
if p_value < alpha:
    print("I will reject the null hypothesis ")
else:
    print("i will accept the null hypothesis")
```
