```python
import numpy as np
import statistics as stats
from scipy.stats import norm
```

```python
# We have taken 10 sample just for an example but *** REMEMBER ***: Our sample should be >30
# The information we have:

sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
population_mean = 170    # avg of population
population_std = 3    # pop. standard deviation 
samp_mean = np.mean(sample)    # avg of values in variable 'sample'
n = len(sample)   # sample size. n=10
```

```python
# Z-test Formula :

# z_value = standardized score (how many std dev away from mean)  
# used to compare sample mean vs population mean in hypothesis testing
z_statistics = (samp_mean - population_mean) / (population_std / np.sqrt(n))    # Formula
z_statistics
```

```python
# p-value formula for 2 tailed test
p_value = 2 * (1 - norm.cdf(abs(z_statistics)))   # cdf = Cumulative Distribution Function. abs = absolute value
```

```python
p_value
```

```python
alpha = 0.05   # significance value

# in p-value it's opposite than z-test. we will reject null hypothesis if p-value is '>=' significance value
# while in z-test we accept null hypothesis when z-test is '<=' significance value
if p_value <= alpha:
    print("I will reject the null hypothesis")
else:
    print("i will accept the null hypothesis ")
```
