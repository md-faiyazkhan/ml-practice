```python
import numpy as np 
from scipy import stats 
```

```python
# for t-test it doesn't matter if the sample size is less than 30 or greater. Best use, if sample size is less than 30.
sample = [172, 174, 168, 169, 171, 173, 175, 170, 169, 172]
sample_mean = np.mean(sample)
sample_std = np.std(sample,ddof = 1)   # ddof = degree of freedom
n = len(sample)
population_mean = 170
```

```python
# t-test Formula
t_stats = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
t_stats
```

```python
# Defining P-value for t-test
p_value = 2 * (1 - stats.t.cdf(abs(t_stats),df = n -1 ))
alpha = 0.05
```

```python
# NOTE: this is opposite to the Z-test. if p-value is less than significance/alpha value then we will reject the 'Null Hypothesis'. If greater, we will accept.
if p_value < alpha:
    print("reject the null hypothesis")
else:
    print("accept the null hypothesis")
```
